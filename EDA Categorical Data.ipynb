```python
# Let's take care of NA's
Input1.isnull().sum().sort_values(ascending=False)
```

```python
#for loop to get the frequency of NA's 
nullcol=['FabricType','Decoration','Material','Pattern Type']
for col in nullcols:
    freq=Input1[col].value_counts(dropna=False)
    print(freq)
```

```python
# Since significant, let's replace NA's with missing for first four columns.
NAImpute=['FabricType', 'Decoration','Material','Pattern Type']
for col in NAImpute:
    Input1[col]=Input1[col].fillna('Missing')
```

```python
# For less significant, let's replace NA's with most frequent value.
for col in Input1.columns:
    Input1[col]=Input1[col].fillna(Input1[col].value_counts().idxmax())
```
