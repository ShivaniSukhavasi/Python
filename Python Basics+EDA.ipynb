```python
#Descriptive statistics- Describe the nature of data only on Numerical data.
#Univariate statistics
# Count, Minimum, Maximum, mesaures of central tendency- Mean, Median, Mode.
# Measures of Dispersion (Scaterring obesrvation around Mean)- Variance & standard deviation
# Measures of location: Quartiles, Percentiles, Decimiles.
# Measures of asymmentry (Normal Distributio to bell curve). 
# Skewness & Kutosis
```

```python
#Mean: Sum of obersvations/Count
# Median: Sort data in ascending order, if odd, midvalue, if even: middle values/2
# Mode: Most frequent value.
# mean and median should have close number, if big difference between mean and median, then data is skewed with outliers.
#mean doesnot always give proper average, so median should be considered for analysis.
```

```python
#Variance- how the data is scattered around mean, if data is closed to mean, variance will be less and vice versa.
#Formula= sum(x-mean)^2/n or (n-1), n is the population, (n-1) is the sample.
# Standard deviation= Square root of Variance
# No fixed range for variance & standard deviation, lower the value the better it is.
```

```python
# Quartiles- Sort data in ascending order. Q1(N/4), Q2(N/50%), Q3(N/75%), Q0 is Minimum, Q4 is Maximum
# Divides data into four equal parts.
```

```python
# deciles- Multiples of 10 & are specific value.
```

```python
#Skewness- Peak of the curve on left is Positive, peak of the curve on right is Negative Skewness.
# Kurtosis: 3
# less than 3 (Negative Kurtosis), peak is Flat & wide. Greater than 3 is peak is narrow.
```

```python
#Data Pre-processing: 95% of time is spent on data cleaning.
# check for Duplicate variable or coloumns.
# Identify based on Data description and delete duplicates.
# Single value columns: All the cells in the columns has same value. Null value coloumn. We must delete this column.
# How to identify, single value column, when mean=median, mean= Quartiles.
```

```python
# Missing value treatment- Identifying missing value on Variable type.
# if mssing values are less than 10% and significant data is available, we can delete the rows that have significant values.
# if misisng values are more than 70% of the column and the variable is deemed to be important, do not impute with mean or median,impute with word 'missing' or 'not available'.
# if missing vales are less than 70% of the column, then mean or median or mode imputation can be done.
# Numeric data: Mean or Median Imputation.
# Non Numeric data: Mode or most frequent Imputation.
# Manual imputation doesn't consider the impact of other variables on the missng values.
# Algortihm based Missing value imputation: This one takes into account of the impact of the variables on missng values and imputes accordingly.
# Eg. MICE- Misisng Imputation using chained equations- Imputes both numeric and non-numeric variables using refression and classification menthods.
# KNN based Imputation: K nearest neighbours Algortim uses distance calculation to impute missimg value. Works well only on numeric data.
# Probablistic PCA- Principal components algorthim.
# tree based Algorithms are alos used for missing value imputations.
```

```python
# Data Transformation: Transforming data structure & type.
# Dependent variable( the value we predict), a numeric and continuous varibale identify skewness,Kutosis and outliers.
# Parametric Algorithms have fundamental assumption of closeness to Normal distribution or bell curve.
# we identify using Histrogram, Box plot, Density curve.
# If there is positive skewness: Logaritmic Transformation or sqaure root.
# Negative Skewness: Exponential or Power transfrmation. 
# if dependet variable(y) is non numericl value, Imbalance data problem.
# Majortity or Minority class: Over sampling (Increase in minority calss) and under sampling (Drecrease in Majority class). SMOKE Algortitm is used for balancing and is a oversampling technique.
# Independent variable (x)- Numeric and continuous variables and there are different mathemaical units like tens, thousands, or discrete variables.
# Scaling must be done. 
# 1. Standard Scaler or Standardization or Z Score= (x-mean)/ Standard Deviation.
# Standard saceler will give you both postive and negative values
# 2. Min max scaler or Normalization or Min Max= (X- X min)/(X max- X min). No negative values.
# 3. Robust scaler= (X- Median(Q2))/(Q3-Q1). Used when there are large amount of outliers im the data. This uses quartles to scale the data.
```
